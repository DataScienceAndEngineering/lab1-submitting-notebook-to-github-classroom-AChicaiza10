```python
resample1['Total Compensation'].hist()
```

```python
def bootstrap_median(original_sample,label,replications):
      just_one_col = original_sample[label]
      medians= np.array([])
      for i in np.arrange(replications):
            bootstrap_sample = just_one_col.sample(500, ignore_index=True)
            resampled_median= np.percentile(bootstrap_sample,50)
            medians = np.append(medians,resampled_median)
      return medians
```

```python
bs_medians = bootstrap_median(full_time,'Total Compensation',500)
```

```python
np.percentile(bs_medians,2.5),np.percentile(bs_medians,97.5)
```
